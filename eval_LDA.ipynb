```python
#define all possible combinations of the classes
classes = ['Medizin', 'Landwirtschaft', 'Umweltwissenschaften', 'Ernährung', 'Rest']
topics = [0.0,1.0,2.0,3.0,4.0]

unique_combinations = []
permut = itertools.permutations(classes, len(topics))

for comb in permut:
    zipped = zip(comb, topics)
    unique_combinations.append(list(zipped))
 
# printing unique_combination list
print(unique_combinations)
```

```python
#get list of the class and predicted topic
pred_averbis = df[['class', 'topic']].values.tolist()
```

```python
def calculate_acc_score(pred_class, combinations):
    result = [] 
    for combi in combinations:
        tmp = {}
        total_count = len(pred_class)
        match_count = 0
        #print(combi)
        for item in pred_class:
            item = tuple(item)
            if item in combi:
                match_count = match_count + 1
        #print(match_count)        
        accuracy = match_count / total_count
        tmp.update({'Combi': combi,'match_count': match_count, 'acc_score': accuracy})
        result.append(tmp)
    return result
```

```python
result = calculate_acc_score(pred_averbis, unique_combinations)
```

```python
pred_averbis_test = df_test[['class', 'topic']].values.tolist()
result_test = calculate_acc_score(pred_averbis_test, unique_combinations)
```

```python
def highest_val(lst, key):
    highest_value = None
    for dict in lst:
        value = dict[key]
        if highest_value is None or value > highest_value:
            highest_value = value
    return highest_value
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

def calculate_f1_score(true_list, predicted_list):
    # Berechnung der true positives (TP), false positives (FP) und false negatives (FN)
    tp = len([1 for true, pred in zip(true_list, predicted_list) if true == pred and true == 1])
    fp = len([1 for true, pred in zip(true_list, predicted_list) if true != pred and true == 0])
    fn = len([1 for true, pred in zip(true_list, predicted_list) if true != pred and true == 1])

    # Berechnung des Precision-Scores
    precision = tp / (tp + fp)

    # Berechnung des Recall-Scores
    recall = tp / (tp + fn)

    # Berechnung des F1-Scores
    f1 = 2 * (precision * recall) / (precision + recall)

    return f1

# Beispielwerte für die true und predicted Listen
true_list = [1, 0, 1, 1, 0]
predicted_list = [1, 0, 0, 1, 1]

# Berechnung des F1-Scores
f1_score = calculate_f1_score(true_list, predicted_list)

print("F1-Score:", f1_score)
```
