```python
def f1_score(true_list, predicted_list):
    # Berechnung der true positives (TP), false positives (FP) und false negatives (FN)
    tp = len([1 for true, pred in zip(true_list, predicted_list) if true == pred and true == 1])
    fp = len([1 for true, pred in zip(true_list, predicted_list) if true != pred and true == 0])
    fn = len([1 for true, pred in zip(true_list, predicted_list) if true != pred and true == 1])

    # Berechnung des Precision-Scores
    precision = tp / (tp + fp)

    # Berechnung des Recall-Scores
    recall = tp / (tp + fn)

    # Berechnung des F1-Scores
    f1 = 2 * (precision * recall) / (precision + recall)

    return precision, recall, f1

# Beispielwerte für die true und predicted Listen
true_list = [1, 0, 1, 1, 0]
predicted_list = [1, 0, 0, 1, 1]

# Berechnung des F1-Scores
f1_score = f1_score(true_list, predicted_list)

print("F1-Score:", f1_score)
```
