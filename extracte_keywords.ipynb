```python
#imports
import pandas as pd
import pickle
from pprint import pprint
from keyword_extraction import DictLU_Extract_Exact
```

```python
from pandarallel import pandarallel  # parallelization
pandarallel.initialize()
```

```python
#read files
df_en = pd.read_csv('/home/ubuntu/ullrich/df_en.csv', sep='|')
df_en = df_en.drop('Unnamed: 0', axis=1)
```

```python
#read files
[dicts_lower,dicts_upper] = pickle.load( open('/home/ubuntu/ullrich/keyword_extraction/pickle/MeSH_dict_english.p', "rb" ) )
DEE=DictLU_Extract_Exact(dicts_upper,dicts_lower)
```

```python
def get_keywords (df, column_name, new_column_name):
    for i, x in enumerate(df[column_name]):
        list_terms = []
        DEE.full(x)
        res = DEE.result
        for k,v in res.items():
            list_terms.append(str(v['term']))
        df.at[i, new_column_name] = ", ".join(list_terms)
    return
```

```python
get_keywords(df_en,'title','keywords_title')
get_keywords(df_en,'abstract','keywords_abs')
```
